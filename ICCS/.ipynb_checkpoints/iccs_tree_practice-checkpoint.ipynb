```python
def iccs_tree_top_province(*context):
    iccs_tree_province = summary_table.groupby(summary_table['location']).emotional_engagement_score.mean().sort_values().head(3)
    iccs_tree_province = iccs_tree_province.to_frame().reset_index()
    location_list = list(iccs_tree_province['location'])
    iccs_tree = pd.DataFrame(columns={ 'parent_id','id','name', 'driver_score'})
    iccs_tree.loc[0, ['name', 'id', 'parent_id', 'driver_score']] = ['ICCS', np.nan,np.nan, np.nan]
    num_list = [1,2,3]
    for i,j in zip(num_list,location_list):
        iccs_tree.loc[i, ['name', 'id', 'parent_id', 'driver_score']] = [j,np.nan,"1",iccs_tree_province.emotional_engagement_score[i-1]]
    for i in range(len(location_list)):
        sample_var = location_list[i]
        print(sample_var)
        vars()[sample_var] = summary_table.loc[summary_table['location']==location_list[i]]
        vars()[sample_var].dropna(axis=1, how='all', inplace=True)
        common_drivers = list(set(drivers_list).intersection(vars()[sample_var].columns))
        vars()[sample_var]= vars()[sample_var][common_drivers].mean().sort_values().head(3)
        vars()[sample_var] = vars()[sample_var].to_frame().reset_index()
        vars()[sample_var].reset_index(inplace=True)
        vars()[sample_var] = vars()[sample_var].drop(vars()[sample_var].columns[[0]], axis=1) 
        vars()[sample_var] = vars()[sample_var].rename(columns = {'index':'name', 0:'driver_score'})
        iccs_tree = iccs_tree.append(vars()[sample_var], ignore_index = True)
        iccs_tree['id'] = list(range(1, iccs_tree.shape[0]+1))
    iccs_tree['parent_id'].iloc[4:7] = '2'
    iccs_tree['parent_id'].iloc[7:10] = '3'
    iccs_tree['parent_id'].iloc[10:13] = '4'
    return iccs_tree
```
